```python
import pandas as pd
from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename)


print("SUS Dataset")
print(df)

sus_columns =[f"Q{1}" for i in range (1,11)]
print(sus_columns)

def calculate_sus_score(row):
    score = 0

    sus_questions = ["Q1","Q2","Q3","Q4","Q5","Q6","Q7","Q8","Q9","Q10"]

    for i, q in enumerate(sus_questions, start=1):
        response = row[q]

        if i % 2 == 1:
            score += response - 1
        else:
            score += 5 - response

    return score * 2.5


df["sus_score"] = df.apply(calculate_sus_score, axis=1)

print("\nIndividual SUS Scores")

if "user" in df.columns:
    print(df[["user", "sus_score"]])
else:
    print(df[["sus_score"]])
```
